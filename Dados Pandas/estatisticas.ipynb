```python
import pandas as pd
```

```python
def carregar_e_limpar(caminho='dados_municipios.csv'):
    dados = pd.read_csv(caminho, encoding='utf-8-sig')

    dados = dados.dropna()

    dados['ano'] = dados['ano'].astype(int)
    dados['total_matriculas'] = dados['total_matriculas'].astype(float)
    dados['pct_privada'] = dados['pct_privada'].astype(float)
    dados['pct_publica'] = dados['pct_publica'].astype(float)

    dados['pct_municipal'] = (
        dados['matriculas_municipal'] / dados['total_matriculas'] * 100
    ).round(1)

    dados['pct_estadual'] = (
        dados['matriculas_estadual'] / dados['total_matriculas'] * 100
    ).round(1)

    dados['pct_federal'] = (
        dados['matriculas_federal'] / dados['total_matriculas'] * 100
    ).round(1)

    return dados
```

```python
dados = carregar_e_limpar()

dados.head()
```

```python
print(dados.shape)

display(
    dados.isnull().sum()
)
```

```python
estatisticas = (
    dados
    .groupby("regiao")
    .agg({
        "pct_privada":"mean",
        "pct_publica":"mean"
    })
)

display(estatisticas)
```

```python
dados.groupby(
    'regiao'
)[
    'pct_publica'
].mean()
```

```python
dados.groupby(
    'regiao'
)[
    'pct_privada'
].mean()
```

```python
evolucao = (
    dados
    .groupby(
        ["regiao","ano"]
    )
    ["pct_privada"]
    .mean()
)

display(evolucao)
```

```python
dados_2024 = (
    dados[
        dados["ano"] == 2024
    ]
)

corr = (
    dados_2024
    .corr(
        numeric_only=True
    )
)

display(corr)
```
